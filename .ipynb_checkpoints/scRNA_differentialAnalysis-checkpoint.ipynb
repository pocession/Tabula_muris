```python
# Packages
import scanpy as sc
import anndata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline

# Dir
dir = '/Users/hsieh/Tabula_muris/'
subset = 'subset'
droplet = 'tabula-muris-senis-droplet-processed-official-annotations.h5ad'
facs = 'tabula-muris-senis-facs-processed-official-annotations.h5ad'
fn_droplet = Path(dir, droplet)
fn_facs = Path(dir, facs)
subset_dir = Path(dir,subset)

# Set variables
method = 'droplet'
# method = 'facs'

# Read data and subsetting it
adata = sc.read_h5ad(fn_droplet)
# adata = sc.read_h5ad(fn_facs)

# Save the raw matrix into raw layer
adata.layers["raw"] = adata.X

# Perform the natural log(X+1) transform and save into log transform layer
adata.layers["log_transformed"] = np.log1p(adata.X)

# View the annotations
adata.obs[:10]
```

```python
# log transform the raw data
# and set the 'raw' attribute to the natural logarithmized data

sc.pp.log1p(adata)
adata.raw = adata
```

```python
# Subset data from large intestine
Tissue = 'Large_Intestine'
adata_subset = adata[adata.obs['tissue'].isin([Tissue])]
```

```python
# Perform the differential analysis

sc.tl.rank_genes_groups(adata, groupby='age', use_raw=True, 
                        method='t-test_overestim_var', n_genes=10) # compute differential expression
sc.pl.rank_genes_groups_tracksplot(adata, groupby='age') # plot the result
```

```python
# Subsetting data
# I want to know how many categories in each column, so I transform the obs to data frame
# and use the pandas function to have this information
# transform the original objects to padas dataframe
df_obs = pd.DataFrame(adata.obs)

# Take a look at the obs data frame
# print(df_obs.head(n=10))
      
# Check how many categories in tissue
print(df_obs['tissue'].cat.categories)
```

```python
# Subset data, only data from intestine are needes
adata_subset = adata[adata.obs['tissue'].isin(['Small-Intestine','Large-Intestine'])]
adata_subset
```

```python
# Perform log transform to the subsetted data
# adata_subset.layers["log_transformed"] = np.log1p(adata_subset.X)
# adata_subset

# Cibersortx does not require data to be log-transformed
adata_subset.layers["raw_data"] = adata_subset.X
adata_subset
```

```python
# transform the raw data to data frame
df_subset = adata_subset.to_df(layer="raw_data")
df_subset
```

```python
# I only need cell type from annotation_lv2
# So let's just keep this column for the subsequent process
df_obs_subset = df_obs['annotation_lv2']
df_subset_new = pd.merge(df_subset, df_obs_subset, left_index=True, right_index=True)
df_subset_new
```

```python
df_subset_new_transposed = df_subset_new.T
df_subset_new_transposed
```

```python
# Separate the count matrix and cell annotation matrix
# Get the count matrix first and save it
sc_matrix = df_subset_new_transposed.drop(index='annotation_lv2')
sc_matrix.to_csv(dir+"scRNA_count.txt", sep="\t")
sc_matrix
```

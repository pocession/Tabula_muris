```python
# transform the raw count matrix to data frame
adata.layers["raw_data"] = adata.X
df_adata = adata.to_df(layer="raw_data")
df_adata.head(10)
```

```python
# save count matrix to csv files
name = method+'_mtx'+'.csv'
df_adata.to_csv(Path(sql_dir,name), sep=',')
```

```python
# Transform annotation to data frame
df_obs = pd.DataFrame(adata.obs)
df_obs.head(10)
```

```python
name=method+'_annot'+'.csv'
df_obs.to_csv(Path(sql_dir,name), sep=',')
```

```python
# Example codes for subsetting data 
# Check how many categories in tissue
print(df_obs['tissue'].cat.categories)
```

```python
# Subset data from large intestine
Tissue = 'Large_Intestine'
adata_subset = adata[adata.obs['tissue'].isin([Tissue])]
adata_subset[[10],[10]]
```

```python
## Perform natural logarithm of one plus the input array (log(x+1))
# adata_subset.layers["log_transformed"] = np.log1p(adata_subset.X)

# Output the raw counts of subsetted data
adata_subset.layers["raw_data"] = adata_subset.X
adata_subset

# transform the raw data to data frame
df_subset = adata_subset.to_df(layer="raw_data")
df_subset
```

```python
# save raw count matrix to csv files
name=method+"_"+Tissue+"_mtx"+".csv"
df_subset.to_csv(Path(sql_dir,name),sep=',')
```

```python
# Subset annotation
Tissue = ('Large_Intestine')
df_obs_subset = df_obs.loc[df_obs['tissue']==Tissue]
```

```python
# save annotation matrix to csv files
name=method+"_"+Tissue+"_annot"+".csv"
df_obs_subset.to_csv(Path(sql_dir,name),sep=',')
```

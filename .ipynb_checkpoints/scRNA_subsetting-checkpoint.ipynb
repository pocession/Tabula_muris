```python
# Packages
import os, sys
import scanpy as sc
import anndata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
%matplotlib inline

# Dir
dir = os.getcwd()
subset = 'subset'
droplet = 'tabula-muris-senis-droplet-processed-official-annotations.h5ad'
facs = 'tabula-muris-senis-facs-processed-official-annotations.h5ad'
fn_droplet = Path(dir, droplet)
fn_facs = Path(dir, facs)
subset_dir = Path(dir,subset)

# Set variables
# method = 'droplet'
method = 'facs'
GroupBy = 'age'

# Read data and subsetting it
# adata = sc.read_h5ad(fn_droplet)
adata = sc.read_h5ad(fn_facs)

# Check how many categories in selected variables
df_obs = pd.DataFrame(adata.obs)
print("-------------------original data--------------------------")
print("-------------------tissue categories----------------------")
print(df_obs['tissue'].cat.categories)
print("-------------------age categories-------------------------")
print(df_obs['age'].cat.categories)

# View the annotations
adata.obs[:10]
```

```python
# Subset data from large intestine
Tissue = ['Large_Intestine']
adata_subset = adata[adata.obs['tissue'].isin(Tissue)]

print("-------------------subsetted data--------------------------")
df_subset_obs = pd.DataFrame(adata_subset.obs)
print("-------------------tissue categories---------------------- ")
print(df_subset_obs['tissue'].cat.categories)
print("-------------------age categories------------------------- ")
print(df_subset_obs['age'].cat.categories)
print("-------------------cell ontology categories--------------- ")
print(df_subset_obs['cell_ontology_class'].cat.categories)
```

```python
# Filter cells
sc.pp.filter_cells(adata_subset, min_genes=200)
sc.pp.filter_genes(adata_subset, min_cells=3)
print(adata_subset)
```

```python
# normalize the expression, the unit is CPM (counts per million)
sc.pp.normalize_total(adata_subset, target_sum=1e6)

# log transform the raw data
# and set the 'raw' attribute to the natural logarithmized data
sc.pp.log1p(adata_subset)

# Freeze the data in log transform value
adata_subset.raw = adata_subset
```

```python
# transform the count matrix to data frame
adata_subset.layers["normalized_data"] = adata_subset.X
df_adata_subset = adata_subset.to_df(layer="normalized_data")

# Take a look at the matrix data
df_adata_subset.head(10)
```

```python
# Transform annotation to data frame
df_obs_subset = pd.DataFrame(adata_subset.obs)
df_obs_subset.head(10)
```

```python
# Save subsetted count matrix and annotation data frame into csv files
tissue_subset = 'LargeIntestine'
annot_name=method+'_'+tissue_subset+'_'+'annot'+'.csv'
mtx_name=method+'_'+tissue_subset+'_'+'mtx'+'.csv'
df_obs_subset.to_csv(Path(subset,annot_name), sep=',')
df_adata_subset.to_csv(Path(subset,mtx_name), sep=',')
```
